```python
# 다양한 타입의 변수 할당
a = 5               # 정수
b = 2.5             # 실수
c = True            # 불리언
d = "파이썬"         # 문자열
e = [1, 2, 3]       # 리스트
f = (4, 5, 6)       # 튜플
g = {"이름": "홍길동", "나이": 30}  # 딕셔너리
h = {1, 2, 2, 3}    # 집합 (중복된 2는 한 번만 저장됨)

print(a, b, c, d)   # 출력: 5 2.5 True 파이썬
print(e)            # 출력: [1, 2, 3]
print(f[1])         # 출력: 5    (튜플 f의 인덱스 1 요소)
print(g["이름"])    # 출력: 홍길동 (딕셔너리 g에서 키 "이름"에 대한 값)
print(h)            # 출력: {1, 2, 3}
```

```python
def add(a, b):
    result = a + b
    return result

def greet(name):
    print(f"안녕하세요, {name}님!")

# 함수 호출 예시
sum_value = add(3, 5)
print("합계:", sum_value)    # 출력: 합계: 8

greet("철수")                # 출력: 안녕하세요, 철수님!
```
