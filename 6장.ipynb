```python
import requests

url = "http://api.example.com/data"
response = requests.get(url)  # GET 방식으로 URL에 요청

if response.status_code == 200:  # 요청 성공 (200 OK)
    data = response.json()       # 응답 내용을 JSON 형식으로 변환
    print(data)
else:
    print("API 호출 실패:", response.status_code)
```
